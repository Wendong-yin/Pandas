## Retrieve A Group with the `.get_group()` Method

```python
fortune = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = fortune.groupby("Sector")
fortune.head(3)
```

通过get_group 得到 dataframe 的子集

```python
sectors.get_group('Technology').head()
```

```python
sectors.get_group('Technology').sort_values('Profits', ascending = False).head()
```

#### 对比传统方法:
.get_group()的优点：
1. 更符合 Python 面向对象的概念，先通过 group by 得到一个特殊的object ，然后通过get方法得到下一个对象

```python
fortune[fortune['Sector'] == 'Technology'].head()
```

我觉得通过中括号来进行截取，感觉很丑

```python
type(sectors.groups)
```

可以通过这个得到 key，然后传入 get_group 方法得到更多的 dataframe

## The `.agg()` Method

```python
fortune = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = fortune.groupby("Sector")
fortune.head(3)
```

```python
sectors["Employees"].mean().head()
```

```python
sectors.agg({"Revenue" : ["sum"],
             "Profits" : "sum",
              "Employees" : "mean"})
```

```python
sectors.agg({"Revenue" : ["sum", "mean"],
             "Profits" : "sum",
              "Employees" : "mean"})
```

```python
sectors.agg(["size", "sum", "mean"])
```

## Iterating through Groups

```python
fortune = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = fortune.groupby("Sector")
fortune.head(3)
```

```python
sectors['Profits'].max()
```

虽然可以得到每个sector的最大值，但我们希望得到这一行的全部信息

先创建一个新的dataframe存储结果

```python
df = pd.DataFrame(columns = fortune.columns)
df # 得到一个 blank 地方
```

用 for 循环遍历sectors中的dataframe

```python
for sector, small_df in sectors:
    highest_revenue_company_in_group = small_df.nlargest(1, "Revenue")
    # 注意 append 没有inplace 功能，必须用一个新的df
    df = df.append(highest_revenue_company_in_group)
```

```python
df
```

```python
cities = fortune.groupby("Location")
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for city, data in cities:
    highest_revenue_in_city = data.nlargest(1, "Revenue")
    df = df.append(highest_revenue_in_city)
```
